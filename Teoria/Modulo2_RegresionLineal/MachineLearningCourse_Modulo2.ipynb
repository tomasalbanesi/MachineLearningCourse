### ¿Que es la regresion lineal?

La regresión lineal es una técnica de análisis de datos que predice el valor de datos desconocidos mediante el uso de otro valor de datos relacionado y conocido. Modela matemáticamente la variable desconocida o dependiente y la variable conocida o independiente como una ecuación lineal. Por ejemplo, supongamos que tiene datos sobre sus gastos e ingresos del año pasado. Las técnicas de regresión lineal analizan estos datos y determinan que tus gastos son la mitad de tus ingresos. Luego calculan un gasto futuro desconocido al reducir a la mitad un ingreso conocido futuro.

### Explicacion de regresion lineal simple

La ecuacion de una recta general tiene la siguiente forma:

\begin{equation}
Y = \beta_0 + \beta_1 X + \epsilon
\end{equation}

donde:

$Y$ es la variable dependiente.

$\beta_0 y \beta_1$ son las constantes que representan el punto de interseccion y la pendiente, respectivamente.

$X$ es la bariable independiente.

$\epsilon$ es el error o funcion de perdida.

### Ejemplo con Python

Se brinda un dataset llamado homeprices.csv en el cual se relacionan distintas areas de casas con sus respectivos precios en dolares estadounidenses. Se debe graficar datos, y utilizando regresion lineal hallar la forma de predecir los precios en base a un area ingresada.

##### Importacion de librerias de Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

##### Importacion de datos CSV

```python
df = pd.read_csv("./data/homeprices.csv")
df
```

##### Grafico scatter

```python
%matplotlib inline
plt.xlabel("area (sqr ft)")
plt.ylabel("price (U$S)")
plt.scatter(df.area, df.price, color="red", marker="+")
```

##### Modelo de regresion lineal

```python
X = df["area"].values.reshape(-1, 1)
Y = df["price"]
reg = linear_model.LinearRegression()
reg.fit(X, Y)
```

##### Predicciones

Como ya tengo definido el modelo de regresion lineal, y se encuentra alimentado con los datos correspondientes, ahora puedo utilizar la prediccion del modelo para estimar los diversos valores de precios en base al area.

```python
x_value = 3300
y_pred = reg.predict([[x_value]])
y_pred
```

```python
coef = reg.coef_
intercept = reg.intercept_
print(f"Coeficiente de pendiente (m): {coef}")
print(f"Interseccion (b): {intercept}")
```

```python
%matplotlib inline
plt.xlabel("area (sqr ft)")
plt.ylabel("price (U$S)")
plt.scatter(df.area, df.price, color="red", marker="+")
plt.plot(df.area, reg.predict(df[["area"]]), color="blue")
```
